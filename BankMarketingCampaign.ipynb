```python
# Drop columns day,month,pdays,previous
data=data.drop(data.columns[[9,10,13,14]], axis = 1) 
```

```python
# Getting dummies
data=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'] ,drop_first=True)
```

```python
# feature scaling all variables to same range
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
columns_to_scale=['age','balance','campaign','duration']
data[columns_to_scale]=sc.fit_transform(data[columns_to_scale])
```

```python
data.head()
```

```python
X=data.drop(['deposit'],axis=1)
y=data['deposit']
```

```python
data.head()
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Implementing XGBoosting Algorithm
import xgboost
from xgboost import XGBClassifier
classifier=xgboost.XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
# Predicting for X_test
y_pred=classifier.predict(X_test)
```

```python
# Getting accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# importing ANN required things 
from keras.models import Sequential
from keras.layers import Dense
```

```python
# initialize ANN
classifier=Sequential()
```

```python
# Add input layer and 1st hidden layer
classifier.add(Dense(units=14,kernel_initializer='uniform',activation='relu',input_dim=28))
```

```python
# add second hidden layer
classifier.add(Dense(units=14,kernel_initializer='uniform',activation='relu'))
```

```python
# Add output layer
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
# compiling ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fitting ANN
classifier.fit(X_train,y_train,batch_size=10,nb_epoch=100)
```

```python
# predicting for test data
y_pred=classifier.predict(X_test)
```

```python
y_pred=(y_pred>0.5)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```
